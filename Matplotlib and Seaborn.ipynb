```python
import numpy as np
x = np.random.randint(0,20,25)
y = np.random.randint(10,30,25)
plt.scatter(x,y,s=np.random.randint(30,100,25)) #s = size 
plt.show()
```

```python
x = np.random.randint(0,20,25)
y = np.random.randint(10,30,25)
plt.scatter(x,y,s=np.random.randint(30,100,25),alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.bar(x,y,color='red',width=0.7)
plt.show()
```

```python
plt.barh(x,y,color='red')
plt.show()
```

```python
import pandas as pd
data = pd.read_csv('mtcars.csv')
data.head(2)
```

```python
#alias
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.lineplot(data['mpg'],data['hp'],data=data)
```

```python
sns.scatterplot(data['mpg'],data['hp'],data=data,alpha=0.5)
```

```python
sns.boxplot(data['mpg'],data=data)
```

```python
print(data['am'].unique())
```

```python
sns.boxplot(data['am'],data['mpg'],data=data)
```

```python
data['am_cat'] = data['am'].apply(lambda x: 'yes' if x==1 else 'No')
```

```python
data.head(1)
```

```python
data['am_cat'].uniqueue()
```

```python
sns.catplot(data['am_cat'],data['wt'],data=data)
```
